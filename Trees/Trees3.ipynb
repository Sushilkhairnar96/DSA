```python
class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right
```

```python
class BT:
    def __init__(self,root=None):
        self.root=TreeNode(root)
        
    def sumRootToLeaf(self, root: TreeNode) -> int:
        
        def path_sum(root,summ):
            if not root:
                return 0
            summ = 2*summ + root.val
            if not root.left and not root.right:
                return summ
            return path_sum(root.left,summ) + path_sum(root.right,summ)
        
        summ=0
        return path_sum(root,summ)
    
    
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
            if not p and not q:
                return True
            if not p or not q:
                return False
            if p.val!=q.val:
                return False
            
            return self.isSameTree(p.left,q.left) and self.isSameTree(p.right,q.right)
    
    
    
    res=float('inf')
    prev=-float('inf')
    
    def minDiffInBST(self, root: TreeNode):  #return the minimum difference between the values of any two different nodes in the tree.
        '''
        The idea is to use an in-order traversal and find the difference between the current node and the node 
        previously checked.
        Think of this question like a variation on "what is the smallest distance between any two values in a sorted array?"
        With a sorted array, all you do is scan through the entire array looking for the absolute minimum difference 
        between two adjacent values since you know, when the array is sorted, that the answer can only be found between 
        those two ADJACENT values.
        
        To scan the BST like a sorted array, just go through IN-ORDER traversal and keep track of the previous node 
        evaluated.
        '''
        if not root:
            return 
        self.minDiffInBST(root.left)
        self.res=min(self.res,root.val-self.prev)
        self.prev=root.val
        self.minDiffInBST(root.right)
        
        return self.res
    
    def sumOfLeftLeaves(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        sum = 0
        if not root:
            return 0
        if root.left and not root.left.left and not root.left.right:
            sum += root.left.val
        sum += self.sumOfLeftLeaves(root.left) + self.sumOfLeftLeaves(root.right)
        return sum
    
    def trimBST(self, root: TreeNode, L: int, R: int) -> TreeNode:
        '''
        If the val of current node is smaller than L, abandon the left sub-tree and trim its right sub-tree
        If the val of current node is greater than R, abandon the right sub-tree and trim its left sub-tree
        Else, recursively trim its left and right sub-tree and return the root
        '''
        if not root:
            return None
        
        root.left=  self.trimBST(root.left,L,R)
        root.right= self.trimBST(root.right,L,R)
        
        if L<= root.val <=R:
            return root
        
        if root.val < L:
            return root.right
        
        if root.val > R :
            return root.left

    def hasPathSum(self, root: TreeNode, sum: int) -> bool:
        '''
        Given a binary tree and a sum, determine if the tree has a root-to-leaf path such that 
        adding up all the values along the path equals the given sum.
        '''
        
        if not root:
            return False
        if not root.left and not root.right and root.val==sum: 
            return True
        return self.hasPathSum(root.left,sum-root.val) or self.hasPathSum(root.right,sum-root.val)
        
        '''
        def find_leaf(summ,root,sum):
            if not root: 
                return False
            if not root.left and not root.right:
                if root.val+summ==sum:
                    return True
                else:
                    return False
            summ+=root.val
            return find_leaf(summ,root.left,sum) or find_leaf(summ,root.right,sum)           
                
        summ=0
        return find_leaf(summ,root,sum)
        '''
    def isBalanced(self, root: TreeNode) -> bool:
        if not root:
            return True

        hleft=self.height(root.left) if root.left else 0
        hright=self.height(root.right) if root.right else 0
        
        if abs(hleft-hright)<=1 and self.isBalanced(root.left) and self.isBalanced(root.right):
            return True
        else:
            return False
        
    def height(self,root):
        if not root:
            return 0
        return max(self.height(root.left),self.height(root.right))+1

    def findTilt(self, root: TreeNode) -> int:              
        if not root: return 0
        left = self.add_BT(root.left)
        right = self.add_BT(root.right)
        return abs(left-right) + self.findTilt(root.left)+ self.findTilt(root.right)

    def add_BT(self,root):
        if not root:return 0
        return root.val + self.add_BT(root.left) + self.add_BT(root.right)
    
    '''
    def findTilt(self, root: TreeNode) -> int:
         def dfs(root):  # get sum of all vals
            if not root:
                return 0
            
            left = dfs(root.left)
            right= dfs(root.right)
            self.tilt += abs(left - right)
            return root.val + left + right
        
        self.tilt = 0
        dfs(root)
        return self.tilt   
    '''
    
    #Two Sum IV - Input is a BST
    '''
    Given a Binary Search Tree and a target number, return true if there exist two elements in the BST 
    such that their sum is equal to the given target.
    '''
    def findTarget(self, root: TreeNode, target: int) -> bool:
        
        self.hash_set=set()
        def helper(root):
            if not root:
                return False
            if target-root.val in self.hash_set:
                return True
            self.hash_set.add(root.val)
            return helper(root.left) or helper(root.right)
        
        return helper(root)  
    
    '''
    Given a non-empty special binary tree consisting of nodes with the non-negative value, where each node in this tree 
    has exactly two or zero sub-node. If the node has two sub-nodes, then this node's value is the smaller value among its
    two sub-nodes. More formally, the property root.val = min(root.left.val, root.right.val) always holds.

    Given such a binary tree, you need to output the second minimum value in the set made of all the nodes' value in the 
    whole tree.

    If no such second minimum value exists, output -1 instead.
    '''
    def findSecondMinimumValue(self, root: TreeNode) -> int:
        self.res=float('inf') 
        root_val=root.val
        def helper(node):
            if not node:
                return 
            if root_val<node.val<self.res:
                self.res=node.val
            helper(node.left)
            helper(node.right)
        helper(root)
        return -1 if self.res==float('inf') else self.res
    '''
    You are given a binary tree in which each node contains an integer value.

    Find the number of paths that sum to a given value.

    The path does not need to start or end at the root or a leaf, but it must go downwards 
    (traveling only from parent nodes to child nodes).
    '''
    count=0
    def pathSum(self, root: TreeNode, summ: int) -> int:
        def helper(root,summ,res):
            if not root:
                return 0
            res + = root.val
            if res==summ:
                self.count+=1
            helper(root.left,summ,res)
            helper(root.right,summ,res)
        
        res=0
        if not root: return 0
        helper(root,summ,res)
        self.pathSum(root.left,summ)
        self.pathSum(root.right,summ)
        return self.count
    
    '''
    Given a binary search tree (BST) with duplicates, find all the mode(s) (the most frequently occurred element) in the 
    given BST.

    The left subtree of a node contains only nodes with keys less than or equal to the node's key.
    The right subtree of a node contains only nodes with keys greater than or equal to the node's key.
    Both the left and right subtrees must also be binary search trees.

    Given BST [1,null,2,2]   => o/p = [2]
    '''
    count=0
    max_count=0
    prev=None
    res=[]       
    
    def findMode(self, root: TreeNode) -> List[int]:     
        self.inorder(root)
        return self.res
    
    def inorder(self,root):
        if not root: return 
        
        self.inorder(root.left)
        
        if self.prev!=root.val:
            self.count=1
        else:
            self.count+=1
        
        if self.count > self.max_count:
            self.res=[]
            self.max_count = self.count
            self.res.append(root.val)
        elif self.count==self.max_count:
            self.res.append(root.val)
        self.prev=root.val
        
        self.inorder(root.right)
        
```
