```python
class TreeNode:
    def __init__(self,val=0,left=None,right=None):
        self.val=val
        self.left=left
        self.right=right
```

```python
class BT:
    def __init__(self,root=None):
        self.root=TreeNode(root)
        
    def sumRootToLeaf(self, root: TreeNode) -> int:
        def bin2dec(s1):
            res=0
            for i in s1:
                res=2*res+int(i)
            return res
        
        def find_leaf(s,root,x):
            if root.left:
                find_leaf(s+str(root.left.val),root.left,x)
            if root.right:
                find_leaf(s+str(root.right.val),root.right,x)
            if not root.left and not root.right:
                x.append(bin2dec(s))
                return sum(x)
            else:
                return sum(x)
        
        s=str(root.val)
        x=[]
        return find_leaf(s,root,x)
    
    
    def isSameTree(self, p: TreeNode, q: TreeNode) -> bool:
        def dfs(p,q):
            if not p and not q:
                return True
            if not p or not q:
                return False
            if p.val!=q.val:
                return False
            return dfs(p.left,q.left) and dfs(p.right,q.right)   
        return dfs(p,q)
    
    res=float('inf')
    prev=-float('inf')
    
    def minDiffInBST(self, root: TreeNode):  #return the minimum difference between the values of any two different nodes in the tree.
        '''
        The idea is to use an in-order traversal and find the difference between the current node and the node 
        previously checked.
        Think of this question like a variation on "what is the smallest distance between any two values in a sorted array?"
        With a sorted array, all you do is scan through the entire array looking for the absolute minimum difference 
        between two adjacent values since you know, when the array is sorted, that the answer can only be found between 
        those two ADJACENT values.
        
        To scan the BST like a sorted array, just go through IN-ORDER traversal and keep track of the previous node 
        evaluated.
        '''
        if not root:
            return 
        self.minDiffInBST(root.left)
        self.res=min(self.res,root.val-self.prev)
        self.prev=root.val
        self.minDiffInBST(root.right)
        
        return self.res
    
    def sumOfLeftLeaves(self, root):
        """
        :type root: TreeNode
        :rtype: int
        """
        sum = 0
        if not root:
            return 0
        if root.left and not root.left.left and not root.left.right:
            sum += root.left.val
        sum += self.sumOfLeftLeaves(root.left) + self.sumOfLeftLeaves(root.right)
        return sum
    
    def trimBST(self, root: TreeNode, L: int, R: int) -> TreeNode:
        '''
        If the val of current node is smaller than L, abandon the left sub-tree and trim its right sub-tree
        If the val of current node is greater than R, abandon the right sub-tree and trim its left sub-tree
        Else, recursively trim its left and right sub-tree and return the root
        '''
        if not root:
            return None
        
        root.left=  self.trimBST(root.left,L,R)
        root.right= self.trimBST(root.right,L,R)
        
        if L<= root.val <=R:
            return root
        
        if root.val < L:
            return root.right
        
        if root.val > R :
            return root.left
```

```python
tree=BT(1)
tree.root.left=TreeNode(0) 
tree.root.right=TreeNode(1) 
tree.root.left.left=TreeNode(0)
tree.root.left.right=TreeNode(1)
tree.root.right.left=TreeNode(0)
tree.root.right.right=TreeNode(1)
'''
Input: [1,0,1,0,1,0,1]
Output: 22
Explanation: (100) + (101) + (110) + (111) = 4 + 5 + 6 + 7 = 22

#           1
#        /    \
#       0      1
#      / \    / \
#     0  1   0   1
'''
print(tree.sumRootToLeaf(tree.root))
```
