```python
def min_heapify(arr,i):
    smallest=i
    left = 2*i+1
    right = 2*i+2
    
    if left < len(arr) and arr[left]<arr[smallest]:
        smallest = left
        
    if right < len(arr) and arr[right]<arr[smallest]:
        smallest = right
        
    if smallest!=i:
        arr[i],arr[smallest] = arr[smallest],arr[i]
        min_heapify(arr,smallest)

def create_heap(arr):
    n=len(arr)
    for i in range(n//2-1,-1,-1):
        min_heapify(arr,i)
        print(arr)
    return arr

arr=[10,30,40,45,60,5,75,90,15]
create_heap(arr)
```

## Time Complexity

https://www.geeksforgeeks.org/time-complexity-of-building-a-heap/

building a heap  = O(n) 

single heapify function cost =O(log(n))

A quick look over the above algorithm suggests that the running time is O(nlg(n)), since each call to Heapify costs O(lg(n)) and Build-Heap makes O(n) such calls.
This upper bound, though correct, is not asymptotically tight.

We can derive a tighter bound by observing that the running time of Heapify depends on the height of the tree ‘h’ (which is equal to log(n), where n is number of nodes) and the heights of most sub-trees are small.
The height ’h’ increases as we move upwards along the tree. create_heap() runs a loop from the index of the last internal node (heapsize/2) with height=1, to the index of root(1) with height = log(n). Hence, Heapify takes different time for each node, which is O(h).



```python
class MinHeap:
    def __init__(self):
        self.arr=[]
        
    def size(self):
        return len(self.arr)
    
    def parent(self,i):
        return (i-1)//2
    
    def left(self,i):
        return 2*i +1
    
    def right(self,i):
        return 2*i +2
    
    def display(self):
        return self.arr
    
    def min_heapify(self,n,i):
        smallest = i
        l = self.left(i)
        r = self.right(i)
        
        if l<n and self.arr[l]<self.arr[smallest]:
            smallest = l
            
        if r<n and self.arr[r]<self.arr[smallest]:
            smallest = r
            
        if smallest!=i:
            self.arr[i],self.arr[smallest]=self.arr[smallest],self.arr[i]
            self.min_heapify(n,smallest)
            
    def create_heap(self):
        n = self.size()
        #start with 1st non-leaf node:
        for i in range(n//2-1,-1,-1):
            self.min_heapify(n,i)
        return self.arr
```
