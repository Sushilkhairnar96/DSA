```python
class Node:
    def __init__(self,data=None):
        self.data=data
        self.next=None
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
        self.size=0
    
    def display(self):
        x=[]
        cur=self.head
        while cur:
            x.append(cur.data)
            cur=cur.next
        print(x)
        
    def get(self,index):
        if index<0 or index>self.size:
            return -1
        cur=self.head
        while index>0:
            cur=cur.next
            index-=1
        return cur.data
        
    def append(self,data):
        new_node=Node(data)
        if not self.head:
            self.head=new_node
            return
        cur=self.head
        while cur.next:
            cur=cur.next
        cur.next=new_node
        self.size+=1
    
    def prepend(self,data):
        new_node=Node(data)
        new_node.next=self.head
        self.head=new_node
        self.size+=1
        
    def insert_after_node(self,prev,data):                 
        cur=self.head
        while cur and cur.data!=prev:
            cur=cur.next
        prev_node=cur
        new_node=Node(data)
        new_node.next=prev_node.next
        prev_node.next=new_node
        self.size+=1
        
    def insertAtIndex(self,index,data) :
        if index<0 or index>self.size:
            return
        elif index==0:
            self.prepend(data)
        elif index == self.size:
            self.append(data)
        else:
            i=0
            new_node=Node(data)
            cur=self.head
            while i<index-1:
                cur=cur.next
                i+=1
            new_node.next=cur.next
            cur.next=new_node
            self.size+=1
    
    def deleteAtHead(self):
        if self.size==1:
            self.head=None
        cur=self.head
        self.head=cur.next
        cur=None
        self.size-=1
    
    def deleteAtTail(self):
        if self.size==1:
            self.head=None
        prev=None
        cur=self.head
        while cur.next:
            prev=cur
            cur=cur.next
        prev.next=None
        self.size-=1
        
    def deleteAtIndex(self,index) :
        if index<0 and index>self.size:
            return 
        if index==0:
            deleteAtHead()
        if index==self.size:
            deleteAtTail()
        else:
            prev=None
            cur=self.head
            i=0
            while cur.next and i<index:
                prev=cur
                cur=cur.next
                i+=1
            prev.next=cur.next
            cur=None
            self.size-=1

    def middleNode(self, head):
        '''
        If object 1 is moving at speed of 2x and object2 at speed of x than object2 will be on halfway when the 
        object1 complete the distance.
        '''
        slow=fast=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        return slow.data
    
    def hasCycle(self, head):
        slow=fast=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
            if slow == fast: return True 
        return False
    
    def getDecimalValue(self, head) -> int:  #Convert Binary Number in a Linked List to Integer
        res = 0
        while head: 
            res = 2*res + head.val 
            head = head.next 
        return res 
    
    def mergeTwoLists1(self, l1, l2):
        head = cur = Node(0)  # a dummy node
        while l1 and l2:
            if l1.val <= l2.val:  # sorted
                cur.next = l1
                l1 = l1.next      
            else:
                cur.next = l2
                l2 = l2.next 
            cur = cur.next 
            
        # when one of them is None, cur should point at the remainder since the linked lists are sorted
        cur.next = l1 if l1 else l2
        return head.next
```

```python
obj=LinkedList()
obj.append(1)
obj.append(2)
obj.append(3)
obj.prepend(4)
obj.display()
obj.insert_after_node(3,5)
obj.display()
obj.get(4)
obj.insertAtIndex(3,6)
obj.display()
obj.deleteAtHead()
obj.display()
obj.deleteAtTail()
obj.display()
obj.deleteAtIndex(2)
obj.display()
obj.append(4)
obj.display()
print(obj.middleNode(obj.head))
```
