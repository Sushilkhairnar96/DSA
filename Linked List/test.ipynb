```python
class Node:
    def __init__(self,val,next=None):
        self.val=val
        self.next=next
```

```python
class LinkedList:
    def __init__(self):
        self.head=None
        self.size=0
    
    def display(self):
        x=[]
        cur= self.head
        while cur!=None:
            x.append(cur.val)
            cur= cur.next
        print("list=",x,"size=",self.size)
    def append(self,val):
        if self.head==None:
            self.head= Node(val)
        else:
            cur= self.head
            while cur.next!=None:
                cur= cur.next
            cur.next=  Node(val)
        self.size+=1
        
    def prepend(self,val):
        if self.head==None:
            self.head= Node(val)
        else:
            new_node = Node(val)
            tmp = self.head
            self.head = new_node
            new_node.next= tmp
        self.size+=1
    
    def insert_after_node(self,prev,val):
        cur = self.head
        while cur and cur.val!= prev:
            cur = cur.next
        prev_node = cur
        new_node = Node(val)
        new_node.next = prev_node.next
        prev_node.next= new_node
        self.size+=1
        
    def insert_at_index(self,index,val):
        cur=self.head
        if index>self.size or index<0:
            return
        elif index==0:
            self.prepend(val)
        elif index==self.size:
            self.append(val)
        else:
            new_node=Node(val)
            i=0
            while i<index-1:
                cur= cur.next
                i+=1
            tmp =cur.next
            cur.next= new_node
            new_node.next= tmp
            self.size+=1
    
    def del_at_head(self):
        if self.head ==None:
            return
        cur =self.head
        self.head = cur.next
        cur =None
        self.size-=1
    
    def del_at_tail(self):
        if self.size==1:
            self.head=None
        prev=None
        cur= self.head
        while cur.next:
            prev = cur
            cur= cur.next
        prev.next=None
        self.size-=1
        
    def del_at_index(self,index):
        cur=self.head
        if index>self.size or index<0:
            return
        elif index==0:
            self.del_at_head()
        elif index==self.size:
            self.del_at_tail()
        else:
            prev=None
            i=0
            while i<index:
                prev=cur
                cur=cur.next
                i+=1
            prev.next=cur.next
            cur=None
            
    def middlenode(self,head):
        slow=fast=head
        while fast and fast.next:
            slow=slow.next
            fast=fast.next.next
        return slow.val
            
        
        

            
```

```python
l1= LinkedList()
for i in range(1,4):
    l1.append(i)
l1.display()
l1.prepend(0)
l1.prepend(6)
l1.display()
l1.insert_after_node(2,5)
l1.display()
l1.insert_at_index(3,8)
l1.display()
l1.middlenode(l1.head)
```
