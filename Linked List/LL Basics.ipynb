```python
class ListNode:
    def __init__(self,val,next=None):
        self.val = val
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        self.size=0
        
    def addAtTail(self,val):
        new_node = ListNode(val)
        if not self.head:
            self.head = new_node
        else:
            cur = self.head
            while cur and cur.next:
                cur=cur.next
            cur.next = new_node
        self.size+=1
        
    def display(self):
        cur = self.head
        x=[]
        while cur:
            x.append(cur.val)
            cur=cur.next
        print(x)
        
    def get(self,index):
        if index<0 or index>=self.size: return -1   #assuming linked list is indexed 0
        if not self.head : return -1
        
        cur = self.head
        while index>0:
            index-=1
            cur=cur.next
        return cur.val 
    
    def addAtHead(self,val):
        new_node = ListNode(val)
        if not self.head : 
            self.head = new_node
        else:
            tmp = self.head
            self.head = new_node
            self.head.next = tmp
        self.size+=1
        
    def insert_after_node(self,prev,val):
        cur = self.head
        new_node = ListNode(val)
        while cur and cur.val!=prev:
            cur = cur.next
        tmp = cur.next
        cur.next = new_node
        new_node.next = tmp
        self.size+=1
        
    def insertAtIndex(self,index,val):
        if index < 0 or index >self.size : 
            print("Invalid index => Index is either exceeding size of L.L or less than 0")
            return 
        if index==0 : self.addAtHead(val)
        elif index==self.size:
            self.addAtTail(val)
        else:
            cur = self.head
            i=0
            new_node = ListNode(val)
            while cur and i!=index-1:   
                cur=cur.next
                i+=1
            tmp=cur.next
            cur.next=new_node
            new_node.next = tmp
            self.size+=1
        
    def deleteAtHead(self):
        if not self.head : return
        self.head=self.head.next
        self.size-=1
    
    def deleteAtTail(self):
        if not self.head : return
        if self.size==1: 
            self.head = None
        else:
            cur = self.head
            prev = None
            while cur.next:
                prev = cur
                cur = cur.next
            prev.next = None
        self.size-=1
        
    def deleteAtIndex(self,index):
        if index<0 or index >=self.size : 
            print("Invalid index => Index is either exceeding size of L.L or less than 0")
            return -1
        if index==0 : 
            self.deleteAtHead()
        elif index==self.size:
            self.deleteAtTail()
        else:
            cur=self.head
            i=0
            while cur and i!=index-1:
                cur=cur.next
                i+=1
            cur.next = cur.next.next
            self.size-=1        
```

```python
obj = LinkedList()
obj.addAtTail(4)
obj.addAtTail(5)
obj.addAtTail(6)
obj.addAtTail(7)
obj.display()
print("value at index 2:",obj.get(2))
obj.addAtHead(2)
obj.addAtHead(1)
obj.display()
obj.insert_after_node(5,8)
obj.display()
obj.insertAtIndex(7,9)
obj.insertAtIndex(0,2)
obj.display()
obj.deleteAtHead()
obj.display()
obj.deleteAtTail()
obj.display()
obj.deleteAtIndex(4)
obj.display()
obj.deleteAtIndex(1)
obj.display()
```
