```python
class Node:
    def __init__(self,val,next=None):
        self.val = val
        self.next = next
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
        self.size=0
        
    def append(self,val):
        if not self.head:
            self.head = Node(val)
        else:
            cur = self.head
            while cur and cur.next:
                cur=cur.next
            cur.next = Node(val)
        self.size+=1
        
    def display(self):
        cur = self.head
        x=[]
        while cur:
            x.append(cur.val)
            cur=cur.next
        print(x)
        
    def get(self,index):
        if index<0 or index>self.size: return -1
        i=0
        cur=self.head
        while i!=index:
            cur = cur.next
            i+=1
        return cur.val
    
    def prepend(self,val):
        new_node = Node(val)
        if not self.head : 
            self.head = new_node
        else:
            tmp = self.head
            self.head = new_node
            self.head.next = tmp
        self.size+=1
        
    def insert_after_node(self,prev,val):
        cur = self.head
        new_node = Node(val)
        while cur and cur.val!=prev:
            cur = cur.next
        tmp = cur.next
        cur.next = new_node
        new_node.next = tmp
        self.size+=1
        
    def insertAtIndex(self,index,val):
        if index<0 or index>self.size: 
            print("Invalid index => Index is either a. exceeding size of L.L or b. less than 0")
            return
        
        new_node=Node(val)
        if index==0:
            self.prepend(val)
        elif index==self.size:
            self.append(val)
        else:
            cur = self.head
            i=0
            while cur and i!=index-1:   #iterate until u reach a node after which u want to insert the element
                cur=cur.next
                i+=1
            tmp=cur.next
            cur.next=new_node
            new_node.next = tmp
        self.size+=1
        
    def deleteAtHead(self):
        if not self.head : return
        cur = self.head
        self.head = cur.next
        self.size-=1
    
    def deleteAtTail(self):
        if not self.head : return
        cur = self.head
        while cur.next.next:   #iterate until u reach second last node and then point it's next to None
            cur= cur.next
        cur.next = None
        self.size-=1
        
    def deleteAtIndex(self,index):
        if index<0 or index>self.size: 
            print("Invalid index => Index is either a. exceeding size of L.L or b. less than 0")
            return
        
        if index==0:
            self.deleteAtHead()
        elif index==self.size:
            self.deleteAtTail()
        else:
            cur=self.head
            i=0
            while cur and i!=index-1:
                cur=cur.next
                i+=1
            cur.next = cur.next.next
        self.size-=1            
```

```python
obj = LinkedList()
obj.append(4)
obj.append(5)
obj.append(6)
obj.append(7)
obj.display()
print("value at index 2:",obj.get(2))
obj.prepend(2)
obj.prepend(1)
obj.display()
obj.insert_after_node(5,8)
obj.display()
obj.insertAtIndex(7,9)
obj.insertAtIndex(0,2)
obj.display()
obj.deleteAtHead()
obj.display()
obj.deleteAtTail()
obj.display()
obj.deleteAtIndex(4)
obj.display()
obj.deleteAtIndex(1)
obj.display()
```
